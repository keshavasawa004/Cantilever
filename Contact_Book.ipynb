```python
import pandas as pd
import os


file_name = "contacts.csv"


if not os.path.exists(file_name):
    df = pd.DataFrame(columns=["Name", "Phone", "Email", "Address"])
    df.to_csv(file_name, index=False)
```

```python
def add_contact(name, phone, email, address):
    df = pd.read_csv(file_name)
    new_contact = pd.DataFrame([{"Name": name, "Phone": phone, "Email": email, "Address": address}])
    df = pd.concat([df, new_contact], ignore_index=True)
    df.to_csv(file_name, index=False)
    print(f"Contact '{name}' added successfully!")


def view_contacts():
    df = pd.read_csv(file_name)
    print(df)


def search_contact(name):
    df = pd.read_csv(file_name)
    result = df[df['Name'].str.lower() == name.lower()]
    print(result if not result.empty else "No contact found.")


def update_contact(name, phone=None, email=None, address=None):
    df = pd.read_csv(file_name)
    index = df.index[df['Name'].str.lower() == name.lower()]
    if not index.empty:
        if phone: df.at[index[0], 'Phone'] = phone
        if email: df.at[index[0], 'Email'] = email
        if address: df.at[index[0], 'Address'] = address
        df.to_csv(file_name, index=False)
        print(f"Contact '{name}' updated successfully!")
    else:
        print("Contact not found.")


def delete_contact(name):
    df = pd.read_csv(file_name)
    df = df[df['Name'].str.lower() != name.lower()]
    df.to_csv(file_name, index=False)
    print(f"Contact '{name}' deleted successfully (if it existed).")
```

```python
while True:
    print("\nðŸ“’ Contact Book")
    print("1. Add Contact")
    print("2. View Contacts")
    print("3. Search Contact")
    print("4. Update Contact")
    print("5. Delete Contact")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == '1':
        name = input("Name: ")
        phone = input("Phone: ")
        email = input("Email: ")
        address = input("Address: ")
        add_contact(name, phone, email, address)

    elif choice == '2':
        view_contacts()

    elif choice == '3':
        name = input("Enter name to search: ")
        search_contact(name)

    elif choice == '4':
        name = input("Enter name to update: ")
        phone = input("New Phone (leave blank if same): ") or None
        email = input("New Email (leave blank if same): ") or None
        address = input("New Address (leave blank if same): ") or None
        update_contact(name, phone, email, address)

    elif choice == '5':
        name = input("Enter name to delete: ")
        delete_contact(name)

    elif choice == '6':
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
